```python
import torch


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Running tensors and pytorch objects on the GPU , thanks to CUDA + NVIDIA hardware + Pytorch working behind the scenes

```python
# setup device agnostic code

device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
# Count number of devices
torch.cuda.device_count()
```

# Putting tensors( and models ) on the GPU
## The reasons we want tensors/models on the GPU is because using GPU results in  faster computations.

```python
# create a tensors (default is on CPU)
tensor = torch.tensor([1,2,3])

# tensor not on GPU
print(tensor, tensor.device)
```

```python
tensor = torch.tensor([1,2,3],device='cpu')

# tensor not on GPU
print(tensor, tensor.device)
```

```python
tensor = torch.tensor([1,2,3],device='cuda')

# tensor not on GPU
print(tensor, tensor.device)
```

```python
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

# Moving tensors back to cpu

```python
tensors_on_cpu = tensor.cpu().numpy()
tensors_on_cpu
```
